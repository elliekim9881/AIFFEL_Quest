```python
import csv

class Aiffel:
    std_dict = {"김건": 0, "김경민": 0, "김소연": 0, "김영신": 0, "박근수": 0, "박영수": 0, "박영준": 0, "서원영": 0, "신지만": 0, "심지안": 0, "이규상": 0,
                 "이윤상": 0, "이진영": 0, "임현석": 0, "전요한": 0, "전재영": 0, "조대희":0}
    dict_num_cnt = 0

    def __init__(self, name, admission_year, num, course): #name 이름, admission_year 입학년도, num 기수, course 과정
        self.name = name
        self.admission_year = admission_year
        self.num = num
        self.course = course
        self.score_sum = 0
        self.score_input = 0
        self.score_penalty = 0

    def generate_id_num(self):
        year = str(self.admission_year)[-2:]    # 연도 마지막 두 자리

        if self.num == "온라인5기":              # 기수
            num = "26"
        else:
            num = "25"

        if self.course == "리서치":              # 과정 종류
            course = "1"
        else:
            course = "2"

        # 가나다순 번호
        list_abc = ["김건", "김경민", "김소연", "김영신", "박근수", "박영수", "박영준", "서원영", "신지만", "심지안", "이규상",
                 "이윤상", "이진영", "임현석", "전요한", "전재영", "조대희"]
        for k in list_abc:
            if k == self.name:
                    num_abc = list_abc.index(k) + 1
                    if num_abc < 10:
                        num_abc = "0" + str(num_abc)
                    else:
                        num_abc = str(num_abc)
        # 학번 생성
        id_num = year + num + course + num_abc  #연도+기수+과정+가나다번호

        for i in self.std_dict:
            if i == self.name:
                self.std_dict[i] = id_num


    def get_id_num(self):
        for i in self.std_dict:
            if self.std_dict[i] != 0:
                self.dict_num_cnt += 1
        print(self.dict_num_cnt)


    def score_chk(self):
        while True:
            plus = int(input("입력할 점수 입력 : "))
            if plus < 1 or plus > 3:
                print("점수는 1에서 3 사이어야 합니다.")
            else:
                self.score_sum += plus
                self.score_input += plus
                break

    def penalty_chk(self):
        i = int(input("차감할 점수 입력 : "))
        self.score_sum -= i
        self.score_penalty += i


    def display_info(self):
        return self.name, self.std_dict[self.name], self.course, self.score_sum, self.score_input, self.score_penalty


    def del_mem_info(self):
        i = input("삭제할 그루의 이름을 입력해주세요")
        while True:
            if i != self.name:
                print("존재하지 않는 그루입니다.")
            else:
                self.std_dict.pop(self.name)
                break

csv_data = []
filename = "core_list.csv"
def made_cvs_list(data, file):
    with open(file, 'w', newline='') as file:
            writer = csv.writer(file)

            writer.writerow(['이름', '학번', '과정종류', '퀘스트 점수 총점', '퀘스트 입력 총점', '퀘스트 penalty 총합']) #헤더 작성

            for i in data:  #리스트를 돌며 csv파일에 입력
                writer.writerow(i)
            print("새로운 코어 명단이 생성되었습니다.")     # 새로운 코어 명단(csv파일)을 생성하기 위해 파일을 열고 writer 객체를 생성, 명단생성완료

a= Aiffel("김건", "2023", "온라인5기", "코어")
b= Aiffel("김경민", "2023", "온라인5기", "코어")
c= Aiffel("김소연", "2023", "온라인5기", "코어")
d= Aiffel("김영신", "2023", "온라인5기", "코어")
e= Aiffel("박근수", "2023", "온라인5기", "코어")
f= Aiffel("박영수", "2023", "온라인5기", "코어")
g= Aiffel("박영준", "2023", "온라인5기", "코어")
h= Aiffel("서원영", "2023", "온라인5기", "코어")
i= Aiffel("신지만", "2023", "온라인5기", "코어")
j= Aiffel("심지안", "2023", "온라인5기", "코어")
k= Aiffel("이규상", "2023", "온라인5기", "코어")
l1= Aiffel("이윤상", "2023", "온라인5기", "코어")
l2= Aiffel("이진영", "2023", "온라인5기", "코어")
l3= Aiffel("임현석", "2023", "온라인5기", "코어")
k1= Aiffel("전요한", "2023", "온라인5기", "코어")
k2= Aiffel("전재영", "2023", "온라인5기", "코어")
k3= Aiffel("조대희", "2023", "온라인5기", "코어")

stu_all_list = [a, b, c, d, e, f, g, h, i, j, k, l1, l2, l3, k1, k2, k3]

a.generate_id_num()
b.generate_id_num()
c.generate_id_num()
d.generate_id_num()
e.generate_id_num()
f.generate_id_num()
g.generate_id_num()
h.generate_id_num()
i.generate_id_num()
j.generate_id_num()
k.generate_id_num()
l1.generate_id_num()
l2.generate_id_num()
l3.generate_id_num()
k1.generate_id_num()
k2.generate_id_num()
k3.generate_id_num()

print(Aiffel.std_dict) #학번 출력
k3.get_id_num() #생성된 학번 개수
k3.score_chk()
k3.penalty_chk()
print(k3.score_sum)

for i in stu_all_list:
   csv_data.append(i.display_info())
made_cvs_list(csv_data, filename)


b.del_mem_info()
d.del_mem_info()
stu_all_list.pop(1)
stu_all_list.pop(3)


# print(Aiffel.std_dict) #학번 출력

```
